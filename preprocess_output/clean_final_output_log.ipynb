```python
import pandas as pd
```

```python
with open("../raw_output/base_output_log_100k.txt", "r") as f:
    raw = f.read().strip()

print(raw)
```

```python
lst = raw.split("\n")
lst
```

```python
y_pred = []
y_true = []
```

```python
tmp = []
for s in lst:
    if s == "":
        continue
    if "===" in s:
        continue
    if "BASE:  torch.Size" in s:
        continue
    if "PATCHES:  torch.Size" in s:
        continue
    
    if "y_true:" in s:
        y_true.append(s[8:].strip("\n"))
        full = " ".join(tmp)
        y_pred.append(full.strip())
        tmp = []
    else:
        tmp.append(s.strip())
```

```python
y_pred
```

```python
y_true
```

```python
len(y_true), len(y_pred)
```

```python
df = pd.DataFrame({
    "y_pred": y_pred,
    "y_true": y_true
})

df
```

```python
df.to_csv("../cleaned_output/base_model_prediction_100k.csv")
```
