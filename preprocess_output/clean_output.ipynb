```python
import pandas as pd
```

```python
with open("../raw_output/finetuned_output_log.txt", "r") as f:
    raw = f.read().strip()

print(raw)
```

```python
while "=" in raw:
    raw = raw.replace("=", "").strip("\n")
```

```python
lst = raw.split("\n")
lst
```

```python
y_pred = []
y_true = []

for s in lst:
    if s == "" or ("BASE:  torch.Size" in s) or ("PATCHES:  torch.Size" in s):
        continue
    elif "y_true" in s:
        y_true.append(s[8:].strip("\n"))
    else:
        y_pred.append(s.strip("\n"))
```

```python
len(y_pred), len(y_true)
```

```python
print(y_pred)
```

```python
print(y_true)
```

```python
df = pd.DataFrame({
    "y_pred": y_pred,
    "y_true": y_true
})

df
```

```python
df.to_csv("../cleaned_output/finetuned_model_prediction_21.csv")
```
